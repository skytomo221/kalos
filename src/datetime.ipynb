```julia
using Dates
df = DateFormat("y-m-d H:M:S")
```

```julia
using CSV, DataFrames, Dates, Plots
histogram(Hour.(DateTime.(DataFrame(CSV.File("tweets.csv"))[!, :DateTime], df)), bins=24, label="@Puomiaviq", xlabel="hour", ylabel="number of tweets", color=RGB(230 / 255, 159 / 255, 0 / 255))
```

```julia
using CSV, DataFrames, Dates, Plots
histogram(Hour.(DateTime.(DataFrame(CSV.File("skytomo221.csv"))[!, :DateTime], df)), bins=24, label="@skytomo221", xlabel="hour", ylabel="number of tweets")
```

```julia
function consecutive(f, A::AbstractVector)
    [f(A[i+1], A[i]) for i = 1:length(A)-1]
end
```

```julia
using CSV, DataFrames, Dates, Plots
histogram(Minute.(consecutive(-, sort(DateTime.(DataFrame(CSV.File("tweets.csv"))[!, :DateTime], df)))), title="Blank time until next tweet", label="@Puomiaviq", xlabel="minutes", ylabel="number of tweets", color=RGB(230 / 255, 159 / 255, 0 / 255))
```

```julia
using CSV, DataFrames, Dates, Plots
histogram(Minute.(consecutive(-, sort(DateTime.(DataFrame(CSV.File("skytomo221.csv"))[!, :DateTime], df)))), title="Blank time until next tweet", label="@skytomo221", xlabel="minutes", ylabel="number of tweets")
```
