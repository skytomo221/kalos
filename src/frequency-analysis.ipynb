```julia
using PyCall
```

```julia
using CSV
using DataFrames
tweets = DataFrame(CSV.File("tweets.csv"))
```

```julia
all_text = join(tweets[!, :Text], " ")
```

```julia
words = split(all_text)
```

```julia
letters = split(all_text, "")
```

```julia
using StatsBase
word_counter = StatsBase.countmap(words)
```

```julia
using StatsBase
letter_counter = StatsBase.countmap(letters)
```

```julia
using Plots
plot(sort(collect(values(word_counter)), rev=true))
```

```julia
using Plots
plot(sort(collect(values(letter_counter)), rev=true), st=:bar)
```
